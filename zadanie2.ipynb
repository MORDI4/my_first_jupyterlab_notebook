```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

prices = [
	(1, 2.12),
	(2, 2.56),
	(3, 3.10),
	(4, 3.16),
	(5, 3.58),
	(6, 5.12),
	(7, 5.16),
	(8, 5.20),
	(9, 4.12),
	(10, 4.10),
	(11, 3.65),
	(12, 4.25),

]
```

```python
df = pd.DataFrame(prices)
data = df[1]/4
data.index = range(1, 13)
```

```python
ax = data.plot(
    kind='line', 
    title='Prices of goods (USD)', 
    linestyle='dashed', 
    color='red', 
    xlabel='month',
    linewidth=3
)
ax.grid(True)
ax.set_ylim(0,2)
ax.set_xlim(0,12.5)
ax.set_xticks(np.arange(0, 13, 2)) 
plt.show()
```
